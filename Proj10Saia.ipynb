```python
# Enable inline plotting and import needed libaries

%matplotlib inline

import pandas as pd
import numpy as np
import datetime as dt
import matplotlib.pyplot as plt
from matplotlib.ticker import AutoMinorLocator
```

```python
# read in csv file then index on Date

data = pd.read_csv('Stock01.csv', index_col = 'Date')
        
# access needed rows and columns
        
data_set1 = data.loc['2006-01-06':'2007-03-20', 'High']
data_set2 = data.loc['2006-01-06':'2007-03-20', 'Low']
data_set3 = data_set1 - data_set2
```

```python
# set up subplots

fig, ax = plt.subplots(3, figsize=(7,3))

# subplot formatting

ax[0].plot(data_set1) 
ax[0].set_ylabel('PRICE')
ax[0].set_title('John Saia')
ax[0].set_xticks(['2006-01-06','2006-06-01', '2006-10-23', '2007-03-20'])
ax[0].grid()
ax[0].legend(['High'], loc="upper center")
ax[0].set_xticklabels([])
ax[0].xaxis.set_minor_locator(AutoMinorLocator(2))

ax[1].plot(data_set2, color = 'orange')
ax[1].set_ylabel('PRICE')
ax[1].set_xticks(['2006-01-06','2006-06-01', '2006-10-23', '2007-03-20'])
ax[1].grid()
ax[1].legend(['Low'],loc="upper center")
ax[1].set_xticklabels([])
ax[1].xaxis.set_minor_locator(AutoMinorLocator(2))

ax[2].plot(data_set3, color = 'green')
ax[2].set_ylabel('PRICE')
ax[2].set_xticks(['2006-01-06','2006-06-01', '2006-10-23', '2007-03-20'])
ax[2].grid()
ax[2].legend(['Difference'], loc="upper left")
ax[2].xaxis.set_minor_locator(AutoMinorLocator(2))
ax[2].set_xlabel('Date')

plt.xticks(rotation=10, ha="right");
plt.tight_layout()
```
